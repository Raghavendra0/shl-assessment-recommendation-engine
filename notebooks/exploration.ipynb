```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("../data/shl_catalog.csv")
df
```

```python
df["combined_text"] = df["skills"] + " " + df["job_roles"]
df[["name", "combined_text"]]
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df["combined_text"])

tfidf_matrix.shape
```

```python
query = "data analyst sql python"
query_vector = vectorizer.transform([query])

scores = cosine_similarity(query_vector, tfidf_matrix)[0]
df["score"] = scores

df.sort_values("score", ascending=False)[["name", "score"]]
```
